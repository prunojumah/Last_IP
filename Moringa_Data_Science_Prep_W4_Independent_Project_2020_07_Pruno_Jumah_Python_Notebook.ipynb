```
#Importing panda and numpy libraries

import pandas as pd
import numpy as np

#loading the csv file on python by first assigning it to a variable
Autolib = '/content/Autolib_dataset (2).csv'

df = pd.read_csv(Autolib)

#viewing the table
df
```

```
#Resizing and cleaning data to remain with what is relevant to our research
df.drop(['Displayed comment','Scheduled at','Geo point','ID','Address'], axis=1, inplace=True)
df.head(10)
```

```
#Capitalizing the column names to enhance uniformity
df.columns = df.columns.str.upper().str.strip().str.replace(' ','_').str.replace('.','')
df.columns
```

```
#Dropping all the entries that are not associated with the city of Paris
df= df.drop(df[(df.CITY !='Paris')].index)

#Displaying the resulting data
df
```

```
#Sorting the dates of the dataset so that it can be in a chronological order
df1 = df.sort_values(by=['POSTAL_CODE','DAY','HOUR','MINUTE'], ascending=True)
df1
```

```
#Creating a column to determine how many shared electric cars (BlueCars) were given out or brought back

df['BLUECARS'] = df.BLUECAR_COUNTER.diff()

#Picking out unique numbers to determine the highest number of cars picked throughout the dataset

df.BLUECARS.unique()
#From the dataset, -7 is the highest number and therefore proceed to determine its equivalent/coresponding time
```

```
Time = df[(df.BLUECARS==7.0)].groupby('HOUR')['HOUR'].count().sort_values(ascending=False)
Time
```

```
#That simply means the highest number in Paris in the month of April 2018 were picked at 1700hrs and 0300hrs
#in 12-hour-clock system, 5pm and 3am
```

```
#Determining the postal code that is the most popular for picking up Blue cars
Time2=df[(df.RENTAL_STATUS=='operational')&(df.BLUECARS)].groupby('POSTAL_CODE')['POSTAL_CODE'].count().sort_values(ascending=False)
Time2.head(20)
```

```
#The most popular postal codes are outlined above.
```

```
#Determining the City that is the most popular for picking up Blue cars
Time2=df1[(df1.RENTAL_STATUS=='operational')&(df.BLUECARS)].groupby('CITY')['CITY'].count().sort_values(ascending=False)
Time2.head(20)
```
